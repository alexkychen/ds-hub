# Clustering Analysis

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alec","Alec","Ben","Charlie","Charlie"],
    "Course":["Math","Bio","Math","Math","Chem"],
    "Grade":["A","B","C","B","C"]
})
df
```

## Preprocessing

### One-hot encoding
Convert categorical to binary data 

```python
# Method 1: Use pandas get_dummies

pd.get_dummies(df, columns=["Course"], dtype=float)
```

```python
# Method 2: Use sklearn OneHotEncoder
# This method help to keep both original and encoded columns
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df[["Course"]])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["Course"]))

pd.concat([df, encoded_df], axis=1)
```

### Ordinal encoding

Convert descriptive variable to ordinal variable

```python
# Method 1: Apply mapping manually
dfc = df.copy()

grade_encoded_map = {"A": 1, "B": 2, "C": 3}

dfc["Grade_encoded"] = dfc["Grade"].map(grade_encoded_map)

dfc
```

```python
# Method 2: Use sklearn OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

ord_encoder = OrdinalEncoder(categories=[["A","B","C"]]) # Make sure it's double squared

dfc["Grade_encoded_2"] = ord_encoder.fit_transform(dfc[["Grade"]]) # Make sure it's double squared

dfc
```
